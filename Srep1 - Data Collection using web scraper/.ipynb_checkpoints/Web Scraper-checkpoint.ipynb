```python
import bs4
import requests 
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
```

```python
def download_image1(url, folder_name, num1):
    
    # write image to file
    reponse = requests.get(url)
    if reponse.status_code == 200:
        with open(os.path.join(folder_name, str(num1) + ".jpg"), 'wb') as file:
            file.write(reponse.content)
    
def download_image2(url, folder_name, num1, num2):
    
    # write image to file
    reponse = requests.get(url)
    if reponse.status_code == 200:
        with open(os.path.join(folder_name, str(num1) + str(num2) + ".jpg"), 'wb') as file:
            file.write(reponse.content)
    
# create a directory to save images
folder_name = 'Images generated' # name whatever you want
if not os.path.isdir(folder_name):
    os.makedirs(folder_name)
```
